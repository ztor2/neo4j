```python
import pandas as pd
from neo4j import GraphDatabase
```

- 생성한 DB 접속 후 connected as > Admin 에서 'server user add'
- ID/PW 설정: jupyter/1234

```python
greeter = GraphDatabase.driver("bolt://localhost:7687", auth=('jupyter', '1234'))
session = greeter.session()
```

```python
nodes = pd.read_csv('data/transport-nodes.csv'); nodes
```

```python
rels = pd.read_csv('data/transport-relationships.csv'); rels
```

#### 노드 추가


```python
def add_nodes(tx, id, latitude, longitude, population):
    tx.run('MERGE (a:Place {id: $id, latitude: $latitude, longitude: $longitude, population: $population})',
            id=id, latitude=float(latitude), longitude=float(longitude), population=int(population))
```

```python
for i in range(len(nodes)):
    session.write_transaction(add_nodes, id=nodes.iloc[i]['id'], latitude=nodes.iloc[i]['latitude'],
                              longitude=nodes.iloc[i]['longitude'], population=nodes.iloc[i]['population'])
```

#### 관계 추가
- 방향성을 넣지 않으면(<>) edge가 양방향의 2개 데이터로 적재되므로 방향성을 부여하는 것이 효율적이다.

```python
def add_rels(tx, src, dst, cost):
    tx.run("MATCH (origin:Place {id: $src}) \
           MATCH (destination:Place {id: $dst}) \
           MERGE (origin)-[r:EROAD {distance: $cost}]->(destination)", src=src, dst=dst, cost=int(cost))
```

```python
for i in range(len(rels)):
    session.write_transaction(add_rels, src=rels.iloc[i]['src'], dst=rels.iloc[i]['dst'], cost=rels.iloc[i]['cost'])
```
